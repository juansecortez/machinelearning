```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

face_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/I.A./haarcascade_fullbody.xml');
img = cv2.imread("/content/drive/MyDrive/I.A./fullbody.jpg")
faces = face_cascade.detectMultiScale(img,scaleFactor=1.1,  minNeighbors=1)
print(faces)
idx=0
for (sx,sy,sw,sh) in faces:
  cv2.rectangle(img,(sx,sy),((sx+sw),(sy+sh)),(0,255,0),2)
  
cv2_imshow(img)
cv2.imwrite('/content/drive/MyDrive/I.A./fullbody_detectadas.jpg',img)
```
